Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def u(n):
    """
    Generates a discrete-time unit step signal u[n].
    """
    # Returns 1.0 where n >= 0, and 0.0 otherwise
    return np.where(n >= 0, 1.0, 0.0)
```

```python
# Define the range of n values being considered
n = np.arange(-10, 10+1, 1) # Array from -10 to 10
```

```python
# Generate the signal
x_n = (n**2)*(u(n-3) - u(n-7))
```

```python
plt.figure(figsize=(10, 6))
plt.stem(n, x_n)
plt.xticks(np.arange(-10, 11, 1))
plt.yticks(np.arange(0, 6+1, 1)**2, rotation=-15)
plt.xlabel('$n$ (samples)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
