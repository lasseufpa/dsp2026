## A) Storage Space

The storage space can be calculated using:

$$
\text{Space (MB)} = \left(\text{transfer rate in MB/s}\right)\times 1800
$$

since $30\ \text{minutes} = 30 \times 60 = 1800\ \text{seconds}$

### 1. PCI
Transfer rate: $266.7\ \text{MB/s}$

Therefore,

$$
\text{Space} = 266.7 \times 1800 = 480060\ \text{MB}
$$

Converting to gigabytes:

$$
480060\ \text{MB} \approx 480.06\ \text{GB}
$$

---

### 2. SATA-300
Transfer rate: $375\ \text{MB/s}$

Therefore,

$$
\text{Space} = 375 \times 1800 = 675000\ = 675\ \text{GB} \text{MB}
$$

---

### 3. USB 2.0
Transfer rate: $60\ \text{MB/s}$

Therefore,

$$
\text{Space} = 60 \times 1800 = 108000\ = 108\ \text{GB} \text{MB}
$$

---

### 4. RS-232
Transfer rate: $0.0288\ \text{MB/s}$

Therefore,

$$
\text{Space} = 0.0288 \times 1800 = 51.84\ \text{MB}
$$

## B) Maximum sampling rate with a 16-bit ADC

If the sampling rate is denoted by $f_s$ (samples/s), then the generated bit rate is:

$$
R_{\text{generated}} = (\text{bits per sample}) \times (\text{samples per second}) = 16f_s
$$

Thus, the maximum sampling rate supported by each interface is:

$$
f_s^{\max} = \frac{R_b}{16}
$$

where $R_b$ is the interface bit rate in bit/s.

### 1. PCI
Bit rate: $R_b = 2133 \times 10^6\ \text{bit/s}$

Therefore,

$$
f_s^{\max} = \frac{2133 \times 10^6}{16} \approx 133.3\ \text{MSamples/s}
$$

---

### 2. SATA-300
Bit rate: $R_b = 3000 \times 10^6\ \text{bit/s}$

Therefore,

$$
f_s^{\max} = \frac{3000 \times 10^6}{16} =  187.5\ \text{MSamples/s}
$$

---

### 3. USB 2.0
Bit rate: $R_b = 480 \times 10^6\ \text{bit/s}$

Therefore,

$$
f_s^{\max} = \frac{480 \times 10^6}{16} = 30\ \text{MSamples/s}
$$

---

### 4. RS-232
Bit rate: $R_b = 0.2304 \times 10^6\ \text{bit/s}$

Therefore,

$$
f_s^{\max} = \frac{0.2304 \times 10^6}{16} = 14.4\ \text{kSamples/s}
$$

##### c) System to sustain 40 MHz sampling and store 3 hours of data

Assume a **16-bit ADC** from last item and 1 channel.

The required data rate is:

$$
R_b = 40 \times 10^6 \times 16 = 640 \times 10^6\ \text{bit/s}
$$

Thus,

$$
R_b = 640\ \text{Mbit/s} = 80\ \text{MB/s}
$$

Therefore, the chosen transfer technology must support at least 80 MB/s:

- **PCI (266.7 MB/s)** can sustain the acquisition
- **SATA-300 (375 MB/s)** can also sustain the acquisition

The acquisition time is:

$$
3\ \text{h} = 3 \times 3600 = 10800\ \text{s}
$$

The total storage space required is:

$$
\text{Space} = 80 \times 10800 = 864000\ \text{MB} = 864\ \text{GB}
$$

##### d) Maximum sampling rate if USB 2.0 must be used

For USB 2.0, the maximum bit rate is: $R_b = 480\ \text{Mbit/s} = 480 \times 10^6\ \text{bit/s}$

Using a 16-bit ADC, the maximum sampling rate is: $f_s^{\max} = \frac{R_b}{16}$

Thus,

$$
f_s^{\max} = \frac{480 \times 10^6}{16} = 30 \times 10^6\ \text{samples/s}
$$

Therefore,

$$
f_s^{\max} = 30\ \text{MSamples/s}
$$

##### e) Example of a Signatec system: maximum throughput for recording hours of data to disk

A direct example is the Signatec DR2800, which is described as being capable of continuous recording up to 2.8 GSPS through PC-to-drive storage streaming.  
In the catalog, recording rates of up to 2800 MB/s are also indicated.

According to the DR2800 configuration table, the maximum sampling rate is $1400\ \text{MSPS}$, the A/D resolution is $8$ bits and the maximum number of channels is $2$.

Source of the table: Signatec Product Catalog (Catalog Revision 2.01, 11/29/2010), section DR2800 System Configurations. 

The total throughput is therefore:

$$
\text{Throughput} = f_s \times N_{\text{bits}} \times N_{\text{channels}}
$$

$$
\text{Throughput} = 1.4 \times 10^9 \times 8 \times 2
= 22.4 \times 10^9\ \text{bit/s}
$$

Thus,

$$
\text{Throughput} = 22.4\ \text{Gbit/s} = 2.8\ \text{GB/s}
$$

To record data for hours at this rate, the storage subsystem must support sustained high-speed writing.
