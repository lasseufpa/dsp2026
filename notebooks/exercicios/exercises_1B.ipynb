## Question 1.13

An analog signal $x(t)$ was periodically sampled with sampling period $T_s = 2\,\mathrm{s}$ to obtain the sampled signal

$$
x_s(t) = 2\delta(t+4) - 5\delta(t-2) + 7\delta(t-6).
$$

The sampling theorem was obeyed.

**a)** What is the analytic expression of $x(t)$?

For ideal sampling,

$$
x_s(t)=\sum_{n=-\infty}^{\infty} x(nT_s)\,\delta(t-nT_s)
$$

with $T_s=2\,\text{s}$.

Given

$$
x_s(t)=2\delta(t+4)-5\delta(t-2)+7\delta(t-6),
$$

the nonzero samples are:

- $x(-4)=2$
- $x(2)=-5$
- $x(6)=7$

and all other samples are zero.

Since the sampling theorem was obeyed, $x(t)$ is reconstructed by sinc interpolation:

$$
x(t)=\sum_{n=-\infty}^{\infty} x(nT_s)\,\mathrm{sinc}\!\left(\frac{t-nT_s}{T_s}\right)
$$

Therefore,

$$
x(t)=2\,\mathrm{sinc}\!\left(\frac{t+4}{2}\right)-5\,\mathrm{sinc}\!\left(\frac{t-2}{2}\right)+7\,\mathrm{sinc}\!\left(\frac{t-6}{2}\right)
$$

where

$$
\mathrm{sinc}(u)=\frac{\sin(\pi u)}{\pi u}.
$$

**b)** Is this the only possible answer? For example, is

$$
2\operatorname{rect}(t+4) - 5\operatorname{rect}(t-2) + 7\operatorname{rect}(t-6)
$$

a possible expression for $x(t)$ given that $T_s = 2\,\mathrm{s}$ and the sampling theorem was obeyed? Why?

Yes, this is the **only possible answer** under the condition that the sampling theorem was obeyed.

So,

$$
2\,\mathrm{rect}(t+4)-5\,\mathrm{rect}(t-2)+7\,\mathrm{rect}(t-6)
$$

is **not** a possible expression for $x(t)$, because the $\mathrm{rect}$ function is **not band-limited**, so it does **not** satisfy the hypothesis of the sampling theorem.

## Question 1.17

The goal is to design a uniform scalar quantizer for a discrete-time signal $x[n]$. It is given that $x[n]$ is approximately uniformly distributed with a mean equal to 2 and a standard deviation equal to 3.

a) Describe a 3-bit quantizer that minimizes the quantization error, explaining your design decisions.

b) Calculate the signal-to-noise ratio (SNR in dB) for the case of an 8-bit quantizer operating with the same input signal.

First thing: the prompt says the signal $x[n]$ is uniform. To design the quantizer, I need to know the signal's boundaries (the limits $a$ and $b$).

Given:

* Mean ($\mu$) = 2

* Standard deviation ($\sigma$) = 3 $\rightarrow$ Variance ($\sigma^2$) = 9

For a uniform distribution:

$$\sigma^2 = \frac{(b-a)^2}{12}$$

$$9 = \frac{(b-a)^2}{12} \implies (b-a)^2 = 108 \implies (b-a) = \sqrt{108} \approx 10.39$$

Since the mean is 2, the interval is centered at 2. So:

* Lower limit ($x_{min}$): $2 - \frac{10.39}{2} = -3.195$

* Upper limit ($x_{max}$): $2 + \frac{10.39}{2} = 7.195$

### a) 3-Bit Quantizer Design

To minimize the error, my design decision is to make the quantizer's dynamic range ($x_{max} - x_{min}$) exactly equal to the signal's width (10.39V). This avoids saturation (clipping) errors and doesn't waste any quantization levels.

Levels ($L$): $2^3 = 8$ levels

Step size ($\Delta$):
$$\Delta = \frac{10.39}{8} \approx 1.3 \text{ V}$$

Configuration: The quantizer will cover the range from -3.195 to 7.195.

* Decision boundaries ($d_i$): Starting at -3.195 and adding 1.3V for each step
* Reconstruction levels ($r_i$): These will be at the midpoint of each step (e.g., the first level is $-3.195 + 0.65 = -2.545\text{V}$)


### b) SNR for 8 Bits
The logic here is the same, just with higher resolution. Since the signal is uniform and the quantizer is designed to perfectly cover the signal's range (as I did in part A), I can use the direct SNR formula for uniform quantization.

For $b = 8$ bits:
The signal power for a uniform signal is $P_s = \frac{\Delta^2 \cdot L^2}{12}$ and the noise power is $P_e = \frac{\Delta^2}{12}$.
The linear ratio is $SNR = L^2 = (2^b)^2 = 2^{2b}$.

Converting to dB:

$$SNR_{dB} = 10 \cdot \log_{10}(2^{2 \cdot 8})$$

$$SNR_{dB} = 160 \cdot \log_{10}(2) \approx 160 \cdot 0.301$$

$$SNR_{dB} \approx 48.16 \text{ dB}$$

Note: I could have used the rule of thumb $6.02 \times 8 \approx 48.1$ dB, which gives the same result.

## Question 1.27

Depict in plots the even and odd parts of

$$
x[n] = \delta[n+1] + 2\delta[n] + 4\delta[n-1].
$$

```python
# eixo n
n = np.arange(-3, 4)

# x[n] = δ[n+1] + 2δ[n] + 4δ[n-1]
x = np.zeros_like(n, dtype=float)
x[n == -1] = 1
x[n == 0] = 2
x[n == 1] = 4

# x[-n]
x_neg = np.zeros_like(n, dtype=float)
x_neg[n == -1] = 4
x_neg[n == 0] = 2
x_neg[n == 1] = 1

# partes par e ímpar
x_even = (x + x_neg) / 2
x_odd = (x - x_neg) / 2

plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.stem(n, x)
plt.title('x[n]')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(n, x_even)
plt.title('Even part')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(n, x_odd)
plt.title('Odd part')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Question 1.28

You do not know much about a signal $x[n]$, but it is given that $x[0] = 3$. Can $x[n]$ be an odd signal? Why?

No, $x[n]$ cannot be an odd signal.

For an odd signal, it must hold that:
$$
x[n] = -x[-n]
$$

At $n=0$:
$$
x[0] = -x[-0] \Rightarrow x[0]=0
$$

But it is given that $x[0]=3 \neq 0$.

Therefore, $x[n]$ is **not odd**.

## Question 1.30

What can be said about the total energy and average power of any periodic signal?

Any periodic signal has:

$$
E_\infty = \infty
$$

because the signal repeats forever, so the total energy diverges.

However, it has finite average power:

$$
P_\infty = \frac{1}{T_0} \int_{T_0} |x(t)|^2 dt
$$

where $T_0$ is the fundamental period.

**a)** $x[n] = u[n] + u[-n-1]$

For any integer $n$:

**b)** $x[n] = \cos\left(\frac{\pi}{8}n\right)$

A discrete-time cosine is periodic if there exists an integer $N_0$ such that

$$
\Omega N_0=2\pi k
$$

$$
\frac{\Omega}{2\pi}=\frac{k}{N_0}
$$

for $k$ and $N_0$ being integers.

Here,
$$
\Omega=\frac{\pi}{8}\text{rad}
$$

So,
$$
\frac{\Omega}{2\pi}=\frac{\frac{\pi}{8}}{2\pi}=\frac{1}{16}
$$

which is a rational number, therefore the signal is periodic.

Thus,

$$
\frac{k}{N_0}=\frac{1}{16}
\Rightarrow N_0=16, \; k=1
$$

Therefore, the signal is periodic with fundamental period $N_0=16$.
