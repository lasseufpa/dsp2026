```python
import numpy as np
import matplotlib.pyplot as plt
```

 Note the several distinct meanings of the word digital depending on the context: an FPGA chip implements digital logic, the signal analyzer has both analog and digital inputs, etc. We want to discuss whether the waveform at a microprocessor data bus pin is a digital signal or not. But according to the definitions adopted in this text, not the nomenclature used in a text dealing with digital electronics. For that, we will assume two cases of a waveform observed with an analog oscilloscope: a) the amplitudes assume only two values: 0 and 5 V, and b) the waveform corresponds to non-ideal pulses, with e. g. non-zero rise and fall times. In these two cases, what are the signal categories according to Table 1.1)?

**Table 1.1:** *Notation used for continuous and discrete-time signals.*

| | Continuous-time, $t \in \mathbb{R}$ | Discrete-time, $n \in \mathbb{Z}$ |
| :--- | :--- | :--- |
| Quantized amplitude | $x_q(t)$ | $x_q[n]$ **(digital)** |
| Not quantized | $x(t)$ **(analog)** | $x[n]$ |

```python
t = np.linspace(0, 4, 1000)
# Generating an ideal square wave that assumes only 0 V and 5 V
y = 5 * (t % 1 < 0.5)

plt.figure(figsize=(8, 4))
plt.plot(t, y, label='Ideal Signal (0 and 5 V)', color='blue', linewidth=2)
plt.title('Signal Representation (Amplitudes of 0 and 5 V)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (V)')
plt.ylim(-1, 6)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# b) Waveform corresponding to non-ideal pulses (non-zero rise and fall times)
t = np.linspace(0, 4, 1000)
y_ideal = 5 * (t % 1 < 0.5)

# Simulating the rise and fall time with a moving average filter (convolution)
window_size = 50
window = np.ones(window_size) / window_size
y_non_ideal = np.convolve(y_ideal, window, mode='same')

plt.figure(figsize=(8, 4))
plt.plot(t, y_non_ideal, label='Non-Ideal Signal (Rise and fall > 0)', color='red', linewidth=2)
plt.title('Signal Representation (Non-Ideal Pulses)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (V)')
plt.ylim(-1, 6)
plt.grid(True)
plt.legend()
plt.show()
```

### Summary of Classifications (Based on Table 1.1)

- **Case a) Ideal pulses (only 0 and 5 V):**
  - **Time:** Continuous ($t \in \mathbb{R}$). The wave is continuous in time.
  - **Amplitude:** Quantized. The signal assumes only two specific levels (0 and 5 V).
  - **Classification:** $x_q(t)$ (Continuous-time, quantized amplitude).

- **Case b) Non-ideal pulses (rise and fall times > 0):**
  - **Time:** Continuous ($t \in \mathbb{R}$).
  - **Amplitude:** Not quantized. It transitions through infinitely many voltage values between 0 V and 5 V.
  - **Classification:** $x(t)$ **(analog)** (Continuous-time, not quantized amplitude).
