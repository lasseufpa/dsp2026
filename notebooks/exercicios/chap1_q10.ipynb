Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x(t):
    """
    Generates the signal depicted in the figure 1.78.
    """
    # Returns 0.75*t where t >= 0 and t < 4, and 0.0 otherwise
    return np.where((t >= 0) & (t < 4.00001), 0.75*t, 0.0)
```

```python
# Plot the signal from the figure 1.78.
t = np.arange(-2, 6+1, 0.1)
x_t = x(t)
plt.plot(t, x_t, label='x(t)')
plt.xticks(np.arange(-2, 6+1, 1))
plt.yticks(np.arange(0, 3+1, 0.5), rotation=-15)
plt.title('Signal x(t)')
plt.xlabel('t (s)')
plt.ylabel('x(t) (V)')
plt.grid()
plt.show()
```
