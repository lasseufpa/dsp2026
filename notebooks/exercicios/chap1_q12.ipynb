Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rect(t):
    """
    Rect function.
    """
    # Returns 1 where |t| <= 0.5, and 0 elsewhere
    return np.where(np.abs(t) <= 0.5, 1.0, 0.0)

def sinc(t):
    """
    Sinc function.
    """
    # Returns np.sinc(t/np.pi) where t != 0, and 1.0 where t == 0
    return np.where(t != 0, np.sinc(t/np.pi), 1.0)
```

```python
Ts = 2
```

a)

```python
t = np.arange(-10, 10+1, 0.1)

z_t = rect((t - 3*Ts)/Ts)

plt.plot(t, z_t, label='z(t)')
plt.xticks(np.arange(-10, 10+1, 1))
plt.xlabel('t')
plt.ylabel('z(t)')
plt.grid()
plt.legend()
plt.show()
```

b)

```python
t = np.arange(-20, 20+1, 0.1) # Extended range to better visualize the signal

x_t = sinc(t/Ts)

plt.plot(t, x_t, label='x(t)')
plt.xticks(np.arange(-20, 20+1, 1))
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.legend()
plt.show()
```

c)

```python
def x_t(t):
    """
    We will need to re-use x_t, but as a function because it will be used as
    building block for other signals. For that, remember that x_t = sinc(t/Ts).
    """
    return sinc(t/Ts)
```

```python
t = np.arange(-20, 20+1, 0.1) # Extended range to better visualize the signal

y_t = 2*x_t(t + Ts) - x_t(t) + 2*x_t(t - Ts)

plt.plot(t, y_t, label='y(t)')
plt.xticks(np.arange(-20, 20+1, 1), rotation=-90)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
plt.show()
```

d)

```python
t = np.arange(-20, 20+1, 0.1) # Extended range to better visualize the signal

v_t = 2*sinc((t + Ts)/Ts) - sinc(t/Ts) + 2*sinc((t - Ts)/Ts)

plt.plot(t, v_t, label='v(t)')
plt.xticks(np.arange(-20, 20+1, 1), rotation=-90)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
plt.show()
```
