An ADC uses $b=8$ 
 bits per sample and sampling frequency $F_s=100$ Msps. a) What is the respective bit rate in bits per second? b) How many megabytes are needed to store 1 hour of a signal digitized with this ADC and stored in a binary “raw” format (without compression or a file header)?

### Resolution
a) Each sample is coding in a 8 bits word length, and we get $100\cdot10^6$ samples per second, then, the bit rate is:
$$R = 8 \cdot 100\cdot10^6 = 800 \cdot10^6 = 800 \text{ Mbps}$$


b) As we know the bit rate $R=800$ Mbps, and 1 hour has 3600 seconds, we compute:
$$\text{Size} = 800 \cdot 10^6 \cdot 3600 $$
$$\text{Size} = 2,88 \text{ Tb}$$
$$\text{Size} = 360  \text{ GB}$$

### B) Maximum sampling rate with a 16-bit ADC

If the sampling rate is denoted by $f_s$ (samples/s), then the generated bit rate is:

$$
R_{\text{generated}} = (\text{bits per sample}) \times (\text{samples per second}) = 16f_s
$$

Thus, the maximum sampling rate supported by each interface is:

$$
f_s^{\max} = \frac{R_b}{16}
$$

where $R_b$ is the interface bit rate in bit/s.

#### 1. PCI
Bit rate: $R_b = 2133 \times 10^6\ \text{bit/s}$

Therefore,

$$
f_s^{\max} = \frac{2133 \times 10^6}{16} \approx 133.3\ \text{MSamples/s}
$$

---

#### 2. SATA-300
Bit rate: $R_b = 3000 \times 10^6\ \text{bit/s}$

Therefore,

$$
f_s^{\max} = \frac{3000 \times 10^6}{16} =  187.5\ \text{MSamples/s}
$$

---

#### 3. USB 2.0
Bit rate: $R_b = 480 \times 10^6\ \text{bit/s}$

Therefore,

$$
f_s^{\max} = \frac{480 \times 10^6}{16} = 30\ \text{MSamples/s}
$$

---

### 4. RS-232
Bit rate: $R_b = 0.2304 \times 10^6\ \text{bit/s}$

Therefore,

$$
f_s^{\max} = \frac{0.2304 \times 10^6}{16} = 14.4\ \text{kSamples/s}
$$

##### c) System to sustain 40 MHz sampling and store 3 hours of data

Assume a **16-bit ADC** from last item and 1 channel.

The required data rate is:

$$
R_b = 40 \times 10^6 \times 16 = 640 \times 10^6\ \text{bit/s}
$$

Thus,

$$
R_b = 640\ \text{Mbit/s} = 80\ \text{MB/s}
$$

Therefore, the chosen transfer technology must support at least 80 MB/s:

- **PCI (266.7 MB/s)** can sustain the acquisition
- **SATA-300 (375 MB/s)** can also sustain the acquisition

The acquisition time is:

$$
3\ \text{h} = 3 \times 3600 = 10800\ \text{s}
$$

The total storage space required is:

$$
\text{Space} = 80 \times 10800 = 864000\ \text{MB} = 864\ \text{GB}
$$

##### d) Maximum sampling rate if USB 2.0 must be used

For USB 2.0, the maximum bit rate is: $R_b = 480\ \text{Mbit/s} = 480 \times 10^6\ \text{bit/s}$

Using a 16-bit ADC, the maximum sampling rate is: $f_s^{\max} = \frac{R_b}{16}$

Thus,

$$
f_s^{\max} = \frac{480 \times 10^6}{16} = 30 \times 10^6\ \text{samples/s}
$$

Therefore,

$$
f_s^{\max} = 30\ \text{MSamples/s}
$$

##### e) Example of a Signatec system: maximum throughput for recording hours of data to disk

A direct example is the Signatec DR2800, which is described as being capable of continuous recording up to 2.8 GSPS through PC-to-drive storage streaming.  
In the catalog, recording rates of up to 2800 MB/s are also indicated.

According to the DR2800 configuration table, the maximum sampling rate is $1400\ \text{MSPS}$, the A/D resolution is $8$ bits and the maximum number of channels is $2$.

Source of the table: Signatec Product Catalog (Catalog Revision 2.01, 11/29/2010), section DR2800 System Configurations. 

The total throughput is therefore:

$$
\text{Throughput} = f_s \times N_{\text{bits}} \times N_{\text{channels}}
$$

$$
\text{Throughput} = 1.4 \times 10^9 \times 8 \times 2
= 22.4 \times 10^9\ \text{bit/s}
$$

Thus,

$$
\text{Throughput} = 22.4\ \text{Gbit/s} = 2.8\ \text{GB/s}
$$

To record data for hours at this rate, the storage subsystem must support sustained high-speed writing.

## Question 6
 Note the several distinct meanings of the word digital depending on the context: an FPGA chip implements digital logic, the signal analyzer has both analog and digital inputs, etc. We want to discuss whether the waveform at a microprocessor data bus pin is a digital signal or not. But according to the definitions adopted in this text, not the nomenclature used in a text dealing with digital electronics. For that, we will assume two cases of a waveform observed with an analog oscilloscope: a) the amplitudes assume only two values: 0 and 5 V, and b) the waveform corresponds to non-ideal pulses, with e. g. non-zero rise and fall times. In these two cases, what are the signal categories according to Table 1.1)?

**Table 1.1:** *Notation used for continuous and discrete-time signals.*

| | Continuous-time, $t \in \mathbb{R}$ | Discrete-time, $n \in \mathbb{Z}$ |
| :--- | :--- | :--- |
| Quantized amplitude | $x_q(t)$ | $x_q[n]$ **(digital)** |
| Not quantized | $x(t)$ **(analog)** | $x[n]$ |

```python
t = np.linspace(0, 4, 1000)
# Generating an ideal square wave that assumes only 0 V and 5 V
y = 5 * (t % 1 < 0.5)

plt.figure(figsize=(8, 4))
plt.plot(t, y, label='Ideal Signal (0 and 5 V)', color='blue', linewidth=2)
plt.title('Signal Representation (Amplitudes of 0 and 5 V)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (V)')
plt.ylim(-1, 6)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# b) Waveform corresponding to non-ideal pulses (non-zero rise and fall times)
t = np.linspace(0, 4, 1000)
y_ideal = 5 * (t % 1 < 0.5)

# Simulating the rise and fall time with a moving average filter (convolution)
window_size = 50
window = np.ones(window_size) / window_size
y_non_ideal = np.convolve(y_ideal, window, mode='same')

plt.figure(figsize=(8, 4))
plt.plot(t, y_non_ideal, label='Non-Ideal Signal (Rise and fall > 0)', color='red', linewidth=2)
plt.title('Signal Representation (Non-Ideal Pulses)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (V)')
plt.ylim(-1, 6)
plt.grid(True)
plt.legend()
plt.show()
```

### Summary of Classifications (Based on Table 1.1)

- **Case a) Ideal pulses (only 0 and 5 V):**
  - **Time:** Continuous ($t \in \mathbb{R}$). The wave is continuous in time.
  - **Amplitude:** Quantized. The signal assumes only two specific levels (0 and 5 V).
  - **Classification:** $x_q(t)$ (Continuous-time, quantized amplitude).

- **Case b) Non-ideal pulses (rise and fall times > 0):**
  - **Time:** Continuous ($t \in \mathbb{R}$).
  - **Amplitude:** Not quantized. It transitions through infinitely many voltage values between 0 V and 5 V.
  - **Classification:** $x(t)$ **(analog)** (Continuous-time, not quantized amplitude).

## Question 7

Draw the graph of  
$x[n] = n^2 \, (\, u[n - 3] - u[n - 7] \,)$

```python
def u(n):
    """
    Generates a discrete-time unit step signal u[n].
    """
    # Returns 1.0 where n >= 0, and 0.0 otherwise
    return np.where(n >= 0, 1.0, 0.0)
```

```python
# Define the range of n values being considered
n = np.arange(-10, 10+1, 1) # Array from -10 to 10
```

```python
# Generate the signal
x_n = (n**2)*(u(n-3) - u(n-7))
```

```python
plt.figure(figsize=(10, 6))
plt.stem(n, x_n)
plt.xticks(np.arange(-10, 11, 1))
plt.yticks(np.arange(0, 6+1, 1)**2, rotation=-15)
plt.xlabel('$n$ (samples)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## Question 12

Carefully draw the signals:

a) $z(t) = \operatorname{rect}\!\left(\dfrac{t - 3T}{T_s}\right)$  

b) $x(t) = \operatorname{sinc}\!\left(\dfrac{t}{T_s}\right)$  

c) $y(t) = 2x(t + T_s) - x(t) + 2x(t - T_s)$  

d) $v(t) = 2\,\operatorname{sinc}\!\left(\dfrac{t + T_s}{T_s}\right) - \operatorname{sinc}\!\left(\dfrac{t}{T_s}\right) + 2\,\operatorname{sinc}\!\left(\dfrac{t - T_s}{T_s}\right)$  

with $T_s = 2$.

```python
def rect(t):
    """
    Rect function.
    """
    # Returns 1 where |t| <= 0.5, and 0 elsewhere
    return np.where(np.abs(t) <= 0.5, 1.0, 0.0)

def sinc(t):
    """
    Sinc function.
    """
    # Returns np.sinc(t/np.pi) where t != 0, and 1.0 where t == 0
    return np.where(t != 0, np.sinc(t/np.pi), 1.0)
```

```python
Ts = 2
```

a)

```python
t = np.arange(-10, 10+1, 0.1)

z_t = rect((t - 3*Ts)/Ts)

plt.plot(t, z_t, label='z(t)')
plt.xticks(np.arange(-10, 10+1, 1))
plt.xlabel('t')
plt.ylabel('z(t)')
plt.grid()
plt.legend()
plt.show()
```

b)

```python
t = np.arange(-20, 20+1, 0.1) # Extended range to better visualize the signal

x_t = sinc(t/Ts)

plt.plot(t, x_t, label='x(t)')
plt.xticks(np.arange(-20, 20+1, 1))
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()
plt.legend()
plt.show()
```

c)

```python
def x_t(t):
    """
    We will need to re-use x_t, but as a function because it will be used as
    building block for other signals. For that, remember that x_t = sinc(t/Ts).
    """
    return sinc(t/Ts)
```

```python
t = np.arange(-20, 20+1, 0.1) # Extended range to better visualize the signal

y_t = 2*x_t(t + Ts) - x_t(t) + 2*x_t(t - Ts)

plt.plot(t, y_t, label='y(t)')
plt.xticks(np.arange(-20, 20+1, 1), rotation=-90)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
plt.show()
```

d)

```python
t = np.arange(-20, 20+1, 0.1) # Extended range to better visualize the signal

v_t = 2*sinc((t + Ts)/Ts) - sinc(t/Ts) + 2*sinc((t - Ts)/Ts)

plt.plot(t, v_t, label='v(t)')
plt.xticks(np.arange(-20, 20+1, 1), rotation=-90)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
plt.show()
```
