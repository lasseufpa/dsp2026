Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Generating signals

```python
x_t = np.zeros(1000)
x_n = np.zeros(10)
```

Setting amplitude

```python
for i in range(400+1):
    x_t[i] = 3

print(x_t)
```

```python
for i in range(5):
    x_n[i] = 3

print(x_n)
```

a) What are the values of x(t) when t = 1.5, 3, and 6?

```python
print(f'x(t=1.5) -> x_t[150] = {x_t[150]}')
print(f'x(t=3) -> x_t[300] = {x_t[300]}')
print(f'x(t=6) -> x_t[600] = {x_t[600]}')
```

b) What are the values of x(n) when n = 1.5, 3, and 6?

```python
try:
    print(f'x(n=1.5) -> x_n[1.5] = {x_n[1.5]}')
except IndexError as e:
    print(f'Error: {e}')
print(f'x(n=3) -> x_n[3] = {x_n[3]}')
print(f'x(n=6) -> x_n[6] = {x_n[6]}')
```

Optional: visualization

```python
n = np.arange(len(x_n))
t = np.arange(len(x_t))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))

ax1.stem(n, x_n)
ax1.set_title('Discrete-Time Signal x[n]')
ax1.set_xlabel('n')
ax1.set_ylabel('x[n]')
ax1.set_xlim(0, 9)
ax1.set_xticks(np.arange(0, 10, 1))
ax1.grid(True)

ax2.plot(t, x_t)
ax2.set_title('Continuous-Time Signal x(t)')
ax2.set_xlabel('t')
ax2.set_ylabel('x(t)')
ax2.set_xlim(0, 900)
ax2.set_xticks(np.arange(0, 901, 100))
ax2.grid(True)

plt.tight_layout()
plt.show()
```
